```python
import pandas as pd
import numpy as np
import plotly.express as px

# import
br = pd.read_csv('Brest.csv')
le = pd.read_csv('Lens.csv')
li = pd.read_csv('Lille.csv')
ly = pd.read_csv('Lyon.csv')
ma = pd.read_csv('Marseille.csv')
mo = pd.read_csv('Monaco.csv')
mn = pd.read_csv('Montpellier.csv')
ni = pd.read_csv('Nice.csv')
pa = pd.read_csv('ParisSG.csv')
re = pd.read_csv('Rennes.csv')

fra = pd.concat([br, le, li, ly, ma, mo, mn, ni, pa, re], ignore_index=True)
fra['From'] = pd.to_datetime(fra['From'], format='%Y-%m-%d')
fra['To'] = pd.to_datetime(fra['To'], format='%Y-%m-%d')
#print(fra)

fra_filtered = fra[fra['From'] >= '2000-07-01']
fig = px.line(fra_filtered, x='From', y='Elo', color='Club', title='France ELO')
fig.update_layout(height=1000)
fig.show()
```
